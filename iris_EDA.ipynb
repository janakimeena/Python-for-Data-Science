### Frequency distribution of SWidth_quarter

```python
pd.crosstab(index=iris['SWidth_quarter'],columns = 'count',dropna = True)
```

```python
pd.crosstab(index=iris['Species'],columns = 'count',dropna = True)
```

### Frequency distribution of SWidth_quarter with respect to different category of flower

```python
pd.crosstab(index=iris['SWidth_quarter'],columns = iris['Species'],dropna = True)
```

### Probability distribution of SWidth_quarter with respect to different category of flower

```python
pd.crosstab(index=iris['SWidth_quarter'],columns = iris['Species'],normalize=True,dropna = True)
```

```python
pd.crosstab(columns=iris['SWidth_quarter'],index = iris['Species'],normalize=True,dropna = True)
```

### Probability of iris having S_Width_Quarter as 1 when type of flower is Iris-setosa or Iris-versicolor or Iris-virginica	is 0.31

```python
pd.crosstab(index=iris['SWidth_quarter'],columns = iris['Species'],normalize=True,dropna = True, margins=True)
```

### Conditional probability is the probability of an event ( A ), given that another event ( B ) has already occurred

### Given the type of SWidth_quarter, probability of different type of iris

```python
pd.crosstab(index=iris['SWidth_quarter'],columns = iris['Species'],normalize='index',margins=True,dropna = True)
```

```python
pd.crosstab(index=iris['SWidth_quarter'],columns = iris['Species'],normalize='columns',margins=True,dropna = True)
```
