## From a CSV File

# Replacing Junk Values as Missing Values

In the iris data, junk values such as "??" and "###" are present in certain columns. They can be converted to missing values by passing them as a list of parameter 'na_values'

```
df = pd.read_csv('Iris_data_sample.csv',index_col=0,na_values=["??","###"])
print(df)
```

## Pandas provide a unique method to retrieve rows from a dataframe using .loc() method

Make the column by which the data is to be filtered as index column while reading the file

```
df = pd.read_csv("Iris_data_sample.csv",index_col='Species')
first = df.loc["Iris-setosa"]
#print(first) #print all rows having Iris-setosa
print(len(first))
```

```
print(list(first.SepalWidthCm))
```

##Retrieving a particular or group of record(s) from DataFrame

.iloc function is used to retrieve nth record from a dataframe

```
row_3 = df.iloc[3]
print(row_3)
```

```
row_3 = df.iloc[3:10]
print(row_3)
```

```
#iloc can also be used for retrieving a column from a dataframe
column_pet = df.loc[:,'SepalLengthCm']
print(column_pet)
```

## Indexing and Selecting Data
To access a scalar value, the fastest way is to use the at and iat methods

```
#at provides label-based scalar lookups
df = pd.read_csv("Iris_data_sample.csv")
print(df.at[5,'SepalLengthCm'])
```

##iat provides integer based lookups

```
df.iat[5,2]
```
