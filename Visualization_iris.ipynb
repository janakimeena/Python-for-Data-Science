```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
iris = pd.read_csv('Iris_data_sample.csv',index_col=0,na_values=['###','??'])
```

```python
iris.dropna(axis=0,inplace=True)
```

### Scatter Plot

```python
plt.scatter(iris["SepalLengthCm"],iris['SepalWidthCm'],c='red')
plt.title('Scatter plot of Sepal Length vs Sepal Width')
plt.xlabel('SepalLength(Cm)')
plt.ylabel('SepalWidth(Cm)')
plt.show()
```

```python
plt.scatter(iris["PetalLengthCm"],iris['PetalWidthCm'],c='blue')
plt.title('Scatter plot of Petal Length vs Petal Width')
plt.xlabel('PetalLength(Cm)')
plt.ylabel('PetalWidth(Cm)')
plt.show()
```

```python
plt.scatter(iris["SepalLengthCm"],iris['PetalLengthCm'],c='orange')
plt.title('Scatter plot of Sepal Length vs Petal Length')
plt.xlabel('SepalLength(Cm)')
plt.ylabel('PetalLength(Cm)')
plt.show()
```

```python
plt.scatter(iris["SepalWidthCm"],iris['PetalWidthCm'],c='orange')
plt.title('Scatter plot of Sepal Width vs Petal Width')
plt.xlabel('SepalWidth(Cm)')
plt.ylabel('PetalWidth(Cm)')
plt.show()
```

```python
plt.scatter(iris["SepalLengthCm"],iris['Species'],c='orange')
plt.title('Scatter plot of Sepal Length vs Species')
plt.xlabel('SepalLength(Cm)')
plt.ylabel('Species')
plt.show()
```

```python
plt.scatter(iris['Species'],iris["SepalWidthCm"],c='violet')
plt.title('Scatter plot of Sepal Length vs Species')
plt.xlabel('SepalLength(Cm)')
plt.ylabel('Species')
plt.show()
```

```python
counts = list(iris['Species'].value_counts())
```

```python
iris_type = list(iris['Species'].unique())
```

```python
index = np.arange(len(iris_type))
```

```python
plt.bar(index,counts,color=['red','blue','cyan'])
plt.title('Bar plot for iris type')
plt.xlabel('Iris types')
plt.ylabel('Frequency')
plt.xticks(index,iris_type,rotation=90)
plt.show()
```
